```python
import requests
import datetime
import json
import os
```

```python
types = ["launches","launchpads","payloads","rockets"]
url = "https://api.spacexdata.com/v4/"
filePath = "/Volumes/esercizio_spacex/bronze/"
curr_date_folder = (datetime.datetime.now()).strftime("%Y-%m-%d")
print(f"Current date: {curr_date_folder}")
```

```python
for t in types:
    response = requests.get(f"{url}{t}").json()
    print(f"Working on API: {url}{t}")
    tmpDirectory = os.path.join(filePath,t,curr_date_folder)
    print(f"Saving into: {tmpDirectory}")
    os.makedirs(tmpDirectory,exist_ok=True)
    with open(os.path.join(tmpDirectory,f"{t}.json"),'w') as f:
        json.dump(response,f)
        print(f"{t.title()} correctly written","\n","-"*100)
```
