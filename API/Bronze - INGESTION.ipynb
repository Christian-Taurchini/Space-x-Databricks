```python
import os
import datetime
from pyspark.sql.functions import current_timestamp
```

```python
types = ["launches","launchpads","payloads","rockets"]
curr_date_folder = (datetime.datetime.now()).strftime("%Y-%m-%d")
filePath = "/Volumes/esercizio_spacex/bronze/"
dbSchema = "esercizio_spacex.bronze.bronze_"
```

```python
for t in types:
    table_name = dbSchema + t
    full_path = os.path.join(filePath,t,curr_date_folder,f"{t}.json")
    print("Working with: ",full_path)

    df = (
        spark.read.text(full_path)
        .withColumnRenamed("value", "raw_payload")
        .withColumn("ingestion_timestamp", current_timestamp())
        .select("ingestion_timestamp", "raw_payload")
    )
    
    df.write \
        .format("delta") \
        .mode("append") \
        .saveAsTable(table_name)
```
